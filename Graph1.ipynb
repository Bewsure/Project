```
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
pd.set_option('display.max_rows',10)
```

```
df=pd.read_csv('https://raw.githubusercontent.com/Aunchittha/CP462PROJECT/master/phuurabpraoychnokhrngkaarthii-yuu-aasaysthaaban.csv')
df
```

```
datamale = df.loc[df['Prefix'] == 'นาย']
datamale
```

```
regiondm = datamale.Region.value_counts()
regiondm
```

```
datafemale = df.loc[df['Prefix'] == 'นาง'] 
datafemale
```

```
regionfdm = datafemale.Region.value_counts()
regionfdm
```

```
datasingle = df.loc[df['Prefix'] == 'นางสาว'] 
datasingle
```

```
newsingle = datasingle.Region.value_counts()
newsingle
```

```
dtfmale = pd.DataFrame(regiondm)
dtfmale
```

```
dtffemale = pd.DataFrame(regionfdm)
dtffemale
```

```
dtfsingle = pd.DataFrame(newsingle)
dtfsingle
```

```
dtfmale = dtfmale.apply(lambda x:((x*100)/34017).astype(float))
dtfmale
```

```
dtffemale = dtffemale.apply(lambda x:((x*100)/34017).astype(float))
dtffemale
```

```
dtfsingle = dtfsingle.apply(lambda x:((x*100)/34017).astype(float))
dtfsingle
```

```
dtfmale.rename(columns={'Region': 'นาย'}, index={'ภาคกรุงเทพฯ ปริมณฑลและตะวันออก': 'ปริมณฑลและตะวันออก'}, inplace=True)
dtfmale
```
